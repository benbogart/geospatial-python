```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

import numpy as np
```

```python
raster_int = np.random.choice(np.arange(0,99), (8,8)).astype('int')
raster_float = np.random.random((8,8))
raster_float.round(2)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 10))

sns.set_context('talk')
sns.heatmap(raster_int, cmap=['w'], annot=True, 
            linewidths=2, linecolor='black', 
            square=True, cbar=False,
            xticklabels=False, yticklabels=False,
            ax=ax1)
ax1.set_title('Raster with integer values')

sns.heatmap(raster_float.round(2), cmap=['w'], annot=True, 
            linewidths=2, linecolor='black', 
            square=True, cbar=False,
            xticklabels=False, yticklabels=False,
            ax=ax2)

ax2.set_title('Raster with float values');
```
